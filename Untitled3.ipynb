```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url=("https://books.toscrape.com/")
soup=BeautifulSoup(requests.get(url).content,"html.parser")
soup.prettify
```

```python
books_list=[]
for book in soup.find_all("li",class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
  name=book.find("h3").text
  price=book.find("p",class_="price_color").text
  rating=book.find("p")["class"][1]
  img="media/cache/2c/da/2cdad67c44b002e7ead0cc35693c0e8b.jpg"+book.find("img",class_="thumbnail")["src"]
  books_list.append([name,price,rating,img])
books_list
```

```python
df=pd.DataFrame(books_list,columns=("name","price","rating","img"))
df
```

```python
df.to_csv("csv_file")
```
